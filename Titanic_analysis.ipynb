```python
import matplotlib.pyplot as plt
```

```python
df["Survived"].value_counts(). plot(kind="bar")
plt.title("Survival count on the Titanic")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of passangers")
plt.show()
```

```python
df["Survived"].value_counts().plot(
    kind="bar",
    color=["red", "green"]
)

plt.title("Titanic Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
df["Pclass"].value_counts()
```

```python
df["Pclass"].value_counts().sort_index().plot(kind="bar",color=["gold", "silver", "brown"])
plt.title("Number of passengers by class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
pd.crosstab(df["Pclass"], df["Survived"])
```

```python
pd.crosstab(df["Pclass"], df["Survived"])
```

```python
pd.crosstab(df["Pclass"], df["Survived"], normalize="index") * 100
```

```python
(pd.crosstab(df["Pclass"], df["Survived"], normalize="index") * 100).plot(kind="bar")

plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Percentage")
plt.legend(["Did Not Survive", "Survived"])

plt.show()
```

```python
pd.crosstab(df["Sex"], df["Survived"], normalize="index") * 100
```

```python
(pd.crosstab(df["Sex"], df["Survived"], normalize="index") * 100).plot(kind="bar")

plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.legend(["Did Not Survive", "Survived"])

plt.show()
```

```python
df_clean = df.copy()
```

```python
df_clean.head()
```

```python
df_clean["Age"] = df_clean["Age"].fillna(df_clean["Age"].mean())
```

```python
df_clean["Embarked"] = df_clean["Embarked"].fillna(df_clean["Embarked"].mode()[0])
```

```python
df_clean = df_clean.drop(columns=["Cabin"])
```

```python
df_clean.isnull().sum()
```

```python
df_clean.to_csv("titanic_clean.csv", index=False)
```
